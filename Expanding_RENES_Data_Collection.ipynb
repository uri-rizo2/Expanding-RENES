   - The incident angle is returned.

# **Function Definition**: *get_tb* and *get_td*

1.In this cell we first define the correction factors defined in the report then define two functions: `get_tb` and `get_td`.
2. The `get_tb` function calculates the clear sky transmittance coefficient for the beam solar radiation for a given row in the `complete_data` DataFrame.
   - It retrieves the altitude value from the `complete_data` DataFrame.
   - The function defines three correction factors: `r0`, `r1`, and `rk`.
   - It calculates the value of `a0` using the altitude and the correction factors.
   - Similarly, it calculates the value of `a1` and `k`.
   - Finally, it calculates the value of `tb` using the formula: `tb = a0 + a1 * (np.exp(-k /math.cos(math.radians(getZen(curr_row)))))`, where `getZen(curr_row)` returns the zenith angle for the corresponding row.
   - The `tb` value is returned.
3. The `get_td` function calculates the clear sky transmittance coefficient for the diffuse solar radiation for a given row in the `complete_data` DataFrame.
   - It calls the `get_tb` function to retrieve the value of `tb` for the corresponding row.
   - The value of `td` is calculated using the formula: `td = 0.271 - (0.294 * get_tb(curr_row))`.
   - The `td` value is returned.
4. The code includes a commented line that demonstrates how to use the `get_tb` function and print its output.






```python
#function that gets clear sky transmittance coefficent for beam solar rad 

#correctioon factors for dfferent climate types
r0 = 1.03
r1 = 1.01
rk = 1.00

def get_tb(curr_row):
  A = complete_data.iloc[curr_row]['Altitude']
  #print(A)
  a0 = r0 * (0.4237 - (0.00821 * ((6 - A)**2)))
  #print(a0)
  a1 = r1 * (0.5055 + (0.00595 * ((6.5 - A)**2)))
  #print(a1)
  k  = rk * (0.2711 + (0.01858 * ((2.5 - A)**2)))
  #print(k)
  tb = a0 + a1 * (np.exp(-k /math.cos(math.radians(getZen(curr_row)))))
  return tb


#function that gets clear sky transmittance coefficent for diffused solar rad 
def get_td(curr_row):
  td = 0.271 - (0.294 * get_tb(curr_row))
  return td

#print (get_tb(54))
```
